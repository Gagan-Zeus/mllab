```python
import csv

num_attributes = 6
a = []

print("\nThe Given Training Data Set\n")

with open('enjoysport.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        a.append(row)
        print(row)
```

```python
print("\nThe initial value of hypothesis:")
hypothesis = ['0'] * num_attributes
print(hypothesis)

# Set hypothesis equal to first positive example (S algorithm start)
for j in range(num_attributes):
    hypothesis[j] = a[0][j]
```

```python
print("\nFind S: Finding a Maximally Specific Hypothesis\n")

for i in range(len(a)):
    if a[i][num_attributes] == 'yes':       # Check if example is positive
        for j in range(num_attributes):
            if a[i][j] != hypothesis[j]:
                hypothesis[j] = '?'
            else:
                hypothesis[j] = a[i][j]

        print(f"For Training instance No:{i} the hypothesis is {hypothesis}")
```

```python
print("\nThe Maximally Specific Hypothesis for the given Training Examples:\n")
print(hypothesis)
```
